```python
import numpy as np
import pandas as pd
```

```python
# read in data
sf_permits = pd.read_csv('C:/Users/Victiny/Python_Project/Data_Cleaning/data_example/Building_Permits.csv')
```

```python
# set seed for reproducibility
np.random.seed(0)
```

```python
"""First look at the data"""
sf_permits.head()
```

```python
# count the number of the missing values
missing_values_count = sf_permits.isnull().sum()
missing_values_count[:10]
```

```python
# number of rows, number of columns
(nrow, ncol) = sf_permits.shape
# number of cells
total_cells = np.product(sf_permits.shape)
# percent of data that is missing
pct_missing = missing_values_count.sum() / total_cells * 100
pct_missing
```

```python
"""Dropping missing values"""
# remove all the rows that contain a missing value
rows_with_na_dropped = sf_permits.dropna()
rows_with_na_dropped 
# this removes all of the adta as every row in the dataset has at least one missing value
```

```python
# remove all columns with at least one missing value
cols_with_na_dropped = sf_permits.dropna(axis=1)
cols_with_na_dropped.head()
```

```python
# how much data we lose?
print('Columns in the original dataset: %d \n' % sf_permits.shape[1])
print("Columns with na's droped: %d" % cols_with_na_dropped.shape[1])
```

```python
"""Filling in missing values automatically"""
subset_sf_permits = sf_permits.loc[:, :].head()
subset_sf_permits
```

```python
# replace NA with 0
subset_sf_permits.fillna(0)
```

```python
# replace NA with the value that comes right after it in the same column
# then replace all the remaining NA with 0
subset_sf_permits.fillna(method='bfill', axis=0).fillna(0)
```
